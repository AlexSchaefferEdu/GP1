```python
# Specify columns to fill in the PCA-KC Data-Assignment.csv file
DTC_Categories = ["Animals / Pets", "City Facilities", "Lights / Signals", "Mowing / Weeds", "Trash / Recycling"]
# Cycle through neighborhoods
for neighborhood in combined_df.NBH_NAME:
# Cycle through specified categories
    for category in DTC_Categories:
        # For each category, neighborhood pair, call the corresponding sub-array from Group2-311DataPreCOVID-Cold Season.csv
        # Compute the average day to close for said sub-array
        fillValue = old_df.loc[((old_df.CATEGORY==category) & (old_df.NEIGH==neighborhood)), "DAYTOCLOSE"].mean()
        # Place the computed value in its corresponding place within PCA-KC Data-Assignment.csv
        combined_df.loc[(combined_df.NBH_NAME == neighborhood), category] = fillValue
```

```python
# Save changes
combined_df.to_csv("PCA-KC Data-Assignment.csv")
```
