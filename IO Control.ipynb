```python
econ = pd.read_excel("data\SocioEconomic.xlsx", sheet_name="SocioEconomic")
econ["Income Level"].unique()
```

```python
h = econ[econ["Income Level"] == "H"] # Boolean selection
m = econ[econ["Income Level"] == "M"]
l = econ[econ["Income Level"] == "L"]
```

```python
def binIncome(cell):
    if str(cell) in list(h["NBH_NAME"].unique()):
        return 'H'
    elif str(cell) in list(m["NBH_NAME"].unique()):
        return 'M'
    elif str(cell) in list(l["NBH_NAME"].unique()):
        return 'L'
    else:
        return np.nan
    
cs["NEIGH-Income"] = cs["NEIGH"].apply(lambda x: binIncome(x))
cs.to_csv("Group2-311DataPreCOVID-Cold Season.csv")
cs[cs["NEIGH-Income"] == 'H'].sort_values(by="CATEGORY").to_csv("data\H-Group2-311DataPreCOVID-Cold Season.csv")
cs[cs["NEIGH-Income"] == 'M'].sort_values(by="CATEGORY").to_csv("data\M-Group2-311DataPreCOVID-Cold Season.csv")
cs[cs["NEIGH-Income"] == 'L'].sort_values(by="CATEGORY").to_csv("data\L-Group2-311DataPreCOVID-Cold Season.csv")
```

```python
H = pd.read_csv("data\H-Group2-311DataPreCOVID-Cold Season.csv")
M = pd.read_csv("data\M-Group2-311DataPreCOVID-Cold Season.csv")
L = pd.read_csv("data\L-Group2-311DataPreCOVID-Cold Season.csv")

ca = pd.read_csv("Categories Assignment.csv")

# For each category label, query the corresponding series from "{}-Group2-311DataPreCOVID-Cold Season.csv", 
# aggregrate sum and mean, then assign to the corresponding place in the Categories Assignment.csv file
for key in list(ca.CATEGORY):
    ca.loc[ca.CATEGORY == key, "Total Time -High"] = H[H.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time High"] = H[H.CATEGORY == key].DAYTOCLOSE.mean()
    
    ca.loc[ca.CATEGORY == key, "Total Time -Moderate"] = M[M.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time Moderate"] = M[M.CATEGORY == key].DAYTOCLOSE.mean()
    
    ca.loc[ca.CATEGORY == key, "Total Time -Low"] = L[L.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time Low"] = L[L.CATEGORY == key].DAYTOCLOSE.mean()
    
# Save changes to the file
ca.loc[:, "CATEGORY":"Avg Time High"].to_csv("Categories Assignment.csv")
```
