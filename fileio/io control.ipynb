```python
econ = pd.read_excel("SocioEconomic.xlsx", sheet_name="SocioEconomic")
econ["Income Level"].unique()
```

```python
h = econ[econ["Income Level"] == "H"] # Boolean selection
m = econ[econ["Income Level"] == "M"]
l = econ[econ["Income Level"] == "L"]

HIN = list(h["NBH_NAME"].unique())  # List of high-income neighborhood labels
MIN = list(m["NBH_NAME"].unique())  
LIN = list(l["NBH_NAME"].unique())
```

```python
def binIncome(cell):
    if str(cell) in HIN:
        return 'H'
    elif str(cell) in MIN:
        return 'M'
    elif str(cell) in LIN:
        return 'L'
    else:
        return np.nan
    
cs["NEIGH-Income"] = cs["NEIGH"].apply(lambda x: binIncome(x))
cs[cs["NEIGH-Income"] == 'H'].to_csv("H-Group2-311DataPreCOVID-Cold Season.csv")
cs[cs["NEIGH-Income"] == 'M'].to_csv("M-Group2-311DataPreCOVID-Cold Season.csv")
cs[cs["NEIGH-Income"] == 'L'].to_csv("L-Group2-311DataPreCOVID-Cold Season.csv")
```

```python
H = pd.read_csv("H-Group2-311DataPreCOVID-Cold Season.csv")
M = pd.read_csv("M-Group2-311DataPreCOVID-Cold Season.csv")
L = pd.read_csv("L-Group2-311DataPreCOVID-Cold Season.csv")

ca = pd.read_csv("Categories Assignment.csv")

# For each category labe, query the corresponding series from "{}-Group2-311DataPreCOVID-Cold Season.csv", 
# aggregrate sum and mean, then assign to the corresponding place in the Categories Assignment.csv file
for key in list(ca.CATEGORY):
    ca.loc[ca.CATEGORY == key, "Total Time -High"] = H[H.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time High"] = H[H.CATEGORY == key].DAYTOCLOSE.mean()
    
    ca.loc[ca.CATEGORY == key, "Total Time -Moderate"] = M[M.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time Moderate"] = M[M.CATEGORY == key].DAYTOCLOSE.mean()
    
    ca.loc[ca.CATEGORY == key, "Total Time -Low"] = L[L.CATEGORY == key].DAYTOCLOSE.sum()
    ca.loc[ca.CATEGORY == key, "Avg Time Low"] = L[L.CATEGORY == key].DAYTOCLOSE.mean()
    
# Save changes to the file
ca.to_csv("Categories Assignment.csv")
```
